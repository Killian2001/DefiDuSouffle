# DicoSon - Jouer un son

## Besoin
Dans le cadre du projet &laquo;Le défi du souffle&raquo; il est avant tout nécessaire de jouer un son au passage de chaque ville. On va s'intéresser ici aux moyens permettant de jouer un son.

## Environnement prélable.
Le programme fonctionnera sur Jupyter, avec le terminal interactif Python `IPython`.
On utilisera un fichier son WAV nommé `test.wav`, qui se trouvera dans le répertoire `sounds` : `sounds/test.wav`.


## Comment jouer un son avec `IPython` ?
`IPython`, en plus des modules Python standards, nous livre un ensemble de modules complémentaires, faits pour fonctionner avec Jupyter : parmi eux se trouve le module `IPython.display`, qui permet d'afficher dans un notebook Jupyter différentes ressources numériques (image, son, vidéo...) *via* un programme Python. Nous allons nous intéresser à l'objet `IPython.display.Audio`.

On importe donc `IPython.display` :

```python
import IPython.display
```

## Qu'est-ce que l'objet `IPython.display.Audio` et comment l'utiliser ?
Il s'agit d'un objet permettant d'afficher un lecteur sonore. On utilise le constructeur de l'objet (méthode spéciale permettant de &laquo;créer&raquo; l'objet), en lui passant un paramètre `filename` (type `str`, chaîne de caractère), contenant l'emplacement du fichier.

```python
IPython.display.Audio(filename='sounds/test.wav')
```

Ici, l'utilisateur doit utiliser manuellement le lecteur multimédia généré par l'objet. Cependant, on souhaiterait jouer automatiquement le son, sans interaction de l'utilisateur.

On utilise pour cela le paramètre `autoplay` (type `bool`, booléen), que l'on met à `True` :

```python
IPython.display.Audio(filename='sounds/test.wav', autoplay=True)
```

Le son est ainsi joué automatiquement.
